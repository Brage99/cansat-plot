```python
import pandas as pd
import simplekml
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

df_radiosonde = pd.read_csv('logs/radiosonde_andoya.csv', sep=',')
df_moist = pd.read_csv('merged_data.txt', sep=',')
```

```python
df_radiosonde['temp_celsius'] = df_radiosonde['temp'] - 273.15
df_radiosonde['elapsed_time'] = df_radiosonde['timestamp'] - 1715079806

df_moist['elapsed_seconds'] = (pd.to_timedelta(df_moist['time']).dt.total_seconds()) - 50738.0
```

```python

plt.plot(df_radiosonde['elapsed_time'], df_radiosonde['alt'])
plt.plot(df_moist['elapsed_seconds'], df_moist['altitude'])
#plt.plot( df_moist['elapsed_seconds'], df_moist['alt'])
plt.xlabel('Time')
plt.ylabel('Altitude')
plt.title('Altitude Radiosonde vs Moist')
plt.show()
```

```python
plt.plot(df_radiosonde['elapsed_time'], df_radiosonde['pressure']*100)
plt.plot(df_moist['elapsed_seconds'], df_moist['pressure'])
plt.xlabel('Time')
plt.ylabel('Pressure (Pa)')
plt.title('Pressure Radiosonde vs Moist')
plt.show()
```

```python
plt.plot(df_radiosonde['elapsed_time'], df_radiosonde['temp_celsius'])
plt.plot(df_moist['elapsed_seconds'], df_moist['temp_ntc'])
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature Radiosonde vs Moist')
plt.show()
```
